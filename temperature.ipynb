```python
#온도 그래프
import pandas as pd

temperature_file = '온도변화_여름_평균.xlsx'
temperature_data = pd.read_excel(temperature_file, index_col=0)

print(temperature_data)
```

```python
# 열 이름 확인
print(temperature_data.columns) 
```

```python
selected_data = temperature_data[['년도', '평균']]
```

```python
selected_data.index
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화
plt.figure(figsize=(10, 6))
sns.lineplot(data=temperature_data, x='년도', y='평균', marker='o', color='red')

# 그래프 제목과 축 레이블 설정
plt.title('평균 온도 변화', fontsize=16)
plt.
plt.xlabel('년도', fontsize=12)
plt.ylabel('평균 온도 (°C)', fontsize=12)
plt.xticks(temperature_data.index, rotation = 45)
plt.yticks([26,26.5,27,27.5,28,28.5,29,29.5,30])
plt.grid(True)

# 그래프 출력
plt.show()
```
