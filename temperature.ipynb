```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Excel 파일 읽기
file_path = './data/평균기온.xlsx'  # 파일 경로
data = pd.read_excel(file_path)
# 한글 폰트 설정 (시스템에 설치된 폰트 경로에 맞게 설정)
plt.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우에서 사용 (리눅스에서는 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 제대로 표시되게 설정

# 2. 데이터 확인
print(data)  # 데이터의 처음 몇 줄 출력
```

```python
# 3. 데이터 시각화
# 예: '연도'별 평균 기온 변화를 선 그래프로 표현한다고 가정
plt.figure(figsize=(12, 6))
plt.plot(data['년도'], data['평균'], marker='o', label='평균기온')

# x축 모든 연도를 표시
plt.xticks(data['년도'], rotation=45)

# 4. 그래프 꾸미기
plt.title('1980~2023년간 평균기온 변화', fontsize=16)
plt.xlabel('년도', fontsize=14)
plt.ylabel('평균기온 (°C)', fontsize=14)

#plt.yticks([10,11,11.5,12,12.5,13,13.5,14,14.5,15,15.5])
# y축 눈금 설정 및 범위 제한

y_ticks = [round(y / 10,1) for y in range(110, 151,2)]  # 11.0 ~ 15.0 범위의 0.1 단위 리스트 생성
plt.yticks(y_ticks)
plt.ylim(11, 15)

plt.grid(True)

plt.legend()
plt.tight_layout()

# 5. 그래프 출력
plt.show()
```
