```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('./data/.xlsx')

#    (     )
plt.rcParams['font.family'] = 'Malgun Gothic'  #   ( 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  #     

# plt.plot(data[''], data[''], marker='o', label='')
years = data.iloc[:,0]
avg_tem = data.iloc[:,1]
plt.figure(figsize=(12, 6))
plt.plot(years,avg_tem)
plt.grid()
plt.legend()
plt.tight_layout()

# x   
plt.xticks(data[''], rotation=45)

# 4.  
plt.title('1980~2023  ', fontsize=14)

# x  
plt.xlim(years.min(), years.max())

plt.xlabel('', fontsize=14)
plt.ylabel(' (C)', fontsize=14)

#  
plt.show()
```
