```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('./data/평균기온.xlsx')

# 한글 폰트 설정 (시스템에 설치된 폰트 경로에 맞게 설정)
plt.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우에서 사용 (리눅스에서는 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 제대로 표시되게 설정

# plt.plot(data['년도'], data['평균'], marker='o', label='평균기온')
years = data.iloc[:,0]
avg_tem = data.iloc[:,1]
plt.figure(figsize=(12, 6))
plt.plot(years,avg_tem)
plt.grid()
plt.legend()
plt.tight_layout()

# x축 모든 연도를 표시
plt.xticks(data['년도'], rotation=45)

# 4. 그래프 꾸미기
plt.title('1980~2023년간 평균기온 변화', fontsize=14)

# x축의 범위 설정
plt.xlim(years.min(), years.max())

plt.xlabel('년도', fontsize=14)
plt.ylabel('평균기온 (°C)', fontsize=14)

# 그래프 출력
plt.show()
```
