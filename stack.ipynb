```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트 설정 (시스템에 설치된 폰트 경로에 맞게 설정)
plt.rcParams['font.family'] = 'Malgun Gothic' 
plt.rcParams['axes.unicode_minus'] = False

data = pd.read_excel('./data/자연재해.xlsx')
data

# 열 데이터 추출
연도 = data['연도'] 
태풍 = data['태풍']    
호우 = data['호우']  
대설 = data['대설'] 

# print(years)
# print(typhoon)
# print(heavy_rain)
# print(heavy_snow)

# 누적 그래프
plt.figure(figsize = (12, 6))
plt.bar(연도, 태풍, label='태풍', color='red')
plt.bar(연도, 호우, bottom = 태풍, label='호우', color='skyblue')  # 태풍 위에 호우
plt.bar(연도, 대설, bottom = 태풍 + 호우, label='대설', color='blue')  # 대설

plt.xlabel('연도')
plt.ylabel('발생 횟수')
plt.title('자연재해 발생횟수 누적 그래프')
plt.legend(loc = 'upper left')  # 범례 생성
plt.tight_layout()
plt.xticks(연도) 
plt.grid() # 그리드
plt.show()  # 그래프 표시
```

```python
# 자연재해
data = pd.read_excel('./data/자연재해.xlsx')

# 열 데이터 추출
연도 = data['연도'] 
태풍 = data['태풍']    
호우 = data['호우']  
대설 = data['대설'] 

# 평균 기온
data2 = pd.read_excel('./data/평균기온.xlsx')

x2 = data2.iloc[20:,0] # 연도 추출
y2 = data2.iloc[20:,1] # 평균 기온 추출

# 그래프 생성
fig, ax1 = plt.subplots(figsize=(12, 6))

# 막대 그래프
ax1.bar(연도, 태풍, label='태풍', color='green')
ax1.bar(연도, 호우, bottom=태풍, label='호우', color='skyblue')  # 태풍 위에 호우
ax1.bar(연도, 대설, bottom=태풍 + 호우, label='대설', color='blue')  # 대설

ax1.set_xlabel('연도')
ax1.set_ylabel('발생 횟수')
ax1.tick_params(axis='y')
ax1.legend(loc='upper left')
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# 꺾은선 그래프 (공유된 x축, 별도의 y축)
ax2 = ax1.twinx()
ax2.plot(x2, y2, color='red', marker='o', label='평균 기온', linewidth=2)
ax2.set_ylabel('평균 기온 (°C)')
ax2.tick_params(axis='y')
ax2.legend(loc='upper right')

# 그래프 출력
plt.tight_layout()
plt.title('기후변화와 자연재해', fontsize=14)
plt.show()
```
