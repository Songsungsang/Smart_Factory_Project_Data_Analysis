```python
import pandas as pd
import matplotlib.pyplot as plt

file = "자연재해_발생횟수(2000~2022).xlsx"
data = pd.read_excel(file, index_col = 0)

# 한글 폰트 설정 (시스템에 설치된 폰트 경로에 맞게 설정)
plt.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우에서 사용 (리눅스에서는 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 제대로 표시되게 설정

# 전처리
x = data.iloc[:2].values
y = data.columns

# 그래프 그리기
plt.figure(figsize=(10, 6))
for disaster in data.index:
    plt.plot(data.columns, data.loc[disaster], marker='o', label=disaster)

# 그래프 제목과 레이블 추가
plt.title('자연재해 연도별 발생 추이 (2000~2022)', fontsize=14)
plt.xlabel('연도', fontsize=12)
plt.ylabel('발생 횟수', fontsize=12)

# 범례 표시
plt.legend(title='재해 유형')

# 그래프 표시
plt.xticks()
plt.grid()
plt.tight_layout()
plt.show()
```
