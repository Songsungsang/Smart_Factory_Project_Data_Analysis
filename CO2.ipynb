```python
import pandas as pd
import matplotlib.pyplot as plt

# 파일 읽기
file1 = './data/CO2.xlsx'
file2 = './data/평균기온.xlsx'

data1 = pd.read_excel(file1)
data2 = pd.read_excel(file2)

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우에서 사용
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 제대로 표시

# 데이터 추출
x1 = data1.columns[1:]  # CO2 연도 추출
y1 = data1.iloc[0, 1:]  # CO2 농도 추출
x2 = data2.iloc[:, 0]   # 평균 기온 연도 추출
y2 = data2.iloc[:, 1]   # 평균 기온 추출

# 연도 데이터 타입 통일 (문자열 -> 정수)
x1 = pd.to_numeric(x1, errors='coerce')
x2 = pd.to_numeric(x2, errors='coerce')

# 공통 연도 찾기
common_years = sorted(set(x1).intersection(set(x2)))

# CO2와 평균 기온 데이터 필터링
y1_filtered = [y1[x1.tolist().index(year)] for year in common_years]
y2_filtered = [y2[x2.tolist().index(year)] for year in common_years]

# 그래프 생성
fig, ax1 = plt.subplots(figsize=(12, 6))

# 막대 그래프
ax1.bar(common_years, y1_filtered, color='skyblue', label='CO2', width=0.4, align='center')
ax1.set_xlabel('연도')
ax1.set_ylabel('CO2 농도 (ppm)')
ax1.tick_params(axis='y')
ax1.legend(loc='upper left')

# 왼쪽 y축 범위 설정
ax1.set_ylim(300, 450)  # y축 범위 설정

# 꺾은선 그래프 (공유된 x축, 별도의 y축)
ax2 = ax1.twinx()
ax2.plot(common_years, y2_filtered, color='red', marker='o', label='평균 기온', linewidth=2)
ax2.set_ylabel('평균 기온 (°C)')
ax2.tick_params(axis='y')
ax2.legend(loc='upper right')

# x축의 범위 설정
# ax1.set_xlim([min(common_years), max(common_years)])  # x축 범위 설정
ax1.set_xticks(common_years)  # x축 레이블 설정
ax1.set_xticklabels(common_years, rotation=45)  # x축 레이블 회전

# 그래프 출력
plt.tight_layout
```
