```python
import pandas as pd
import matplotlib.pyplot as plt

#  
file1 = './data/CO2.xlsx'
file2 = './data/.xlsx'

data1 = pd.read_excel(file1)
data2 = pd.read_excel(file2)

#   
plt.rcParams['font.family'] = 'Malgun Gothic'  #  
plt.rcParams['axes.unicode_minus'] = False  #    

#  
x1 = data1.columns[1:]  # CO2  
y1 = data1.iloc[0, 1:]  # CO2  
x2 = data2.iloc[:, 0]   #    
y2 = data2.iloc[:, 1]   #   

#     ( -> )
x1 = pd.to_numeric(x1, errors='coerce')
x2 = pd.to_numeric(x2, errors='coerce')

#   
common_years = sorted(set(x1).intersection(set(x2)))

# CO2    
y1_filtered = [y1[x1.tolist().index(year)] for year in common_years]
y2_filtered = [y2[x2.tolist().index(year)] for year in common_years]

#  
fig, ax1 = plt.subplots(figsize=(12, 6))

#  
ax1.bar(common_years, y1_filtered, color='skyblue', label='CO2', width=0.4, align='center')
ax1.set_xlabel('')
ax1.set_ylabel('CO2  (ppm)')
ax1.tick_params(axis='y')
ax1.legend(loc='upper left')

#  y  
ax1.set_ylim(300, 450)  # y  

#   ( x,  y)
ax2 = ax1.twinx()
ax2.plot(common_years, y2_filtered, color='red', marker='o', label=' ', linewidth=2)
ax2.set_ylabel('  (C)')
ax2.tick_params(axis='y')
ax2.legend(loc='upper right')

# x  
# ax1.set_xlim([min(common_years), max(common_years)])  # x  
ax1.set_xticks(common_years)  # x  
ax1.set_xticklabels(common_years, rotation=45)  # x  

#  
plt.tight_layout
```
