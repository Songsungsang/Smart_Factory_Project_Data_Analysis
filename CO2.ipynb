```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  
file = 'CO2.xlsx'
data = pd.read_excel(file)

#    (     )
plt.rcParams['font.family'] = 'Malgun Gothic'  #   ( 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  #     

#  
x = data.columns[1:] #  
y = data.iloc[0, 1:] # CO2  

# #  
# plt.figure(figsize=(10, 6))
# plt.bar(x, y, color='skyblue', label='CO2 Levels (ppm)')

# #  
# plt.title(' CO2 Levels (ppm)')
# plt.xlabel('')
# plt.ylabel('CO2  (ppm)')
# plt.grid()
# plt.legend()
# plt.xticks()
# plt.yticks()
# plt.tight_layout()

#  
fig, ax1 = plt.subplots(figsize=(12, 6))

# y   (     )
plt.ylim(350, 440)

#  
ax1.bar(x, y, color='skyblue', label='CO2')
ax1.set_xlabel('')
ax1.set_ylabel('CO2  (ppm)')
ax1.tick_params(axis='y')
ax1.legend(loc='upper left')

#   ( x,  y)
ax2 = ax1.twinx()
ax2.plot(x, y, color='red', marker='o', label=' ')
ax2.set_ylabel(' ')
ax2.tick_params(axis='y')
ax2.legend(loc='upper right')

#  
# plt.title(' CO2 Levels (ppm)')
plt.show()
```
