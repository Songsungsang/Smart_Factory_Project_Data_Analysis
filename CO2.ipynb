```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 파일 읽기
file = 'CO2.xlsx'
data = pd.read_excel(file)

# 한글 폰트 설정 (시스템에 설치된 폰트 경로에 맞게 설정)
plt.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우에서 사용 (리눅스에서는 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 제대로 표시되게 설정

# 데이터 추출
x = data.columns[1:] # 연도 추출
y = data.iloc[0, 1:] # CO2 농도 추출

# # 그래프 생성
# plt.figure(figsize=(10, 6))
# plt.bar(x, y, color='skyblue', label='CO2 Levels (ppm)')

# # 그래프 커스터마이징
# plt.title('연도별 CO2 Levels (ppm)')
# plt.xlabel('연도')
# plt.ylabel('CO2 농도 (ppm)')
# plt.grid()
# plt.legend()
# plt.xticks()
# plt.yticks()
# plt.tight_layout()

# 그래프 생성
fig, ax1 = plt.subplots(figsize=(12, 6))

# y축 범위 조정 (데이터 최소값 기준으로 하단 여백 축소)
plt.ylim(350, 440)

# 막대 그래프
ax1.bar(x, y, color='skyblue', label='CO2')
ax1.set_xlabel('연도')
ax1.set_ylabel('CO2 농도 (ppm)')
ax1.tick_params(axis='y')
ax1.legend(loc='upper left')

# 꺾은선 그래프 (공유된 x축, 별도의 y축)
ax2 = ax1.twinx()
ax2.plot(x, y, color='red', marker='o', label='평균 기온')
ax2.set_ylabel('평균 기온')
ax2.tick_params(axis='y')
ax2.legend(loc='upper right')

# 그래프 출력
# plt.title('연도별 CO2 Levels (ppm)')
plt.show()
```
